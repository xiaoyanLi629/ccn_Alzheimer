```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import pprint
import numpy as np
import pandas as pd
```

```python
data_frame = pd.read_csv('oasis_cross-sectional.csv')
data_frame.drop('Delay', axis=1, inplace=True)
data_frame = data_frame.dropna()
data_frame[:3]
```

```python
# correlation matrix of the data
cor_matrix = np.corrcoef(data_frame[['Age', 'Educ', 'eTIV', 'nWBV', 'ASF', 'MMSE', 'CDR', 'SES']].T)
#show the correlation matrix as image
plt.imshow(cor_matrix, cmap='hot', interpolation='nearest')
# add a colorbar
plt.colorbar()
plt.show()
```
